```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris= sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].value_counts()
```

```python
pair = sns.PairGrid(iris)
pair.map(plt.scatter)
```

```python
pair2 = sns.PairGrid(iris)
pair2.map_diag(plt.hist)
pair2.map_upper(plt.scatter)
pair2.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris, hue='species')
```
