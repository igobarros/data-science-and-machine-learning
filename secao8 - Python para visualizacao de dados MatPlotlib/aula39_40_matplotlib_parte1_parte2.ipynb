```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x * x
y
```

```python
plt.plot(x, y, color='r')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
plt.plot(x, y, 'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(x, y, 'g*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_ylabel('Eixo Y')
axes1.set_title('Título')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_ylabel('Eixo Y')
axes1.set_title('Título')
axes2.plot(y, x)
```

```python
fig1, ax1 = plt.subplots() # ax --> Eixo
ax1.plot(x, x**3, 'b--')
ax1.plot(x, x**4, 'g--')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Título')
```

```python
fig2, ax2 = plt.subplots(nrows=1, ncols=2)
```

```python
ax2
```

```python
for axe in ax2:
    axe.plot(x, y, 'g')
    axe.set_title('Título')
fig2
```

```python
ax2[0].plot(x, y, 'b')
ax2[0].set_title('Título A')

ax2[1].plot(x, y, 'r')
ax2[1].set_title('Título B')
fig2
```

```python
fig3, ax3 = plt.subplots(nrows=5, ncols=5)
fig3.tight_layout()
```

```python
fig4 = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig5, ax5 = plt.subplots(figsize=(6, 4), dpi=80)

ax5.plot(x, y, 'b', label='x, y')
ax5.plot(y, x, 'r--', label='y, x')
ax5.set_title('Título')
ax5.legend(loc=4) # loc define em qual posição as legendas irá se possicionar
```

```python
fig5.savefig('imagem1.jpg') # salva o gráfico em diversos formatos
```
